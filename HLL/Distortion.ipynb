```python
#Distortion effect
"""
Implementacion del efecto de distorsion en un archivo de audio.

"""

import numpy as np
import matplotlib.pyplot as plot
from scipy.io import wavfile
```

```python
#funciones:
def dibujar(x,y):
  #dibujamos la senoidal.
  plot.plot(x,y)
  plot.title('Seno')
  plot.xlabel('Tiempo')
  plot.ylabel('Amplitud')
  plot.grid(True, which='both')
  plot.axhline(y=0, color='k')
  plot.xscale("linear")
  plot.xlim(0,2)
  # mostrar la grafica.
  plot.show()


```

```python
from IPython.display import Audio, display
import numpy as np
#Muestra de audio original.
sample_rate, data = wavfile.read('./audios/masterofpuppets-mono.wav')
#Duracion del archivo original de audio.
duration = 5.0 
time = np.arange(0, duration, 1/sample_rate)
silence = np.zeros(sample_rate*int(duration))

#Reproducir audio original.
display(Audio('./audios/SINE.wav', autoplay=False))

#dibujar audio
plot.plot(time, data[:len(time)])

#Ajustar escala en X\
plot.xlim(1, 1.03455)
plot.ylim(-16000,16000)
plot.xlabel('Tiempo [s]')
plot.ylabel('Amplitud')
plot.title('Grabacion 2. Sin alteraciones') 
plot.show()


```

## Resultado del efecto

### Grabacion 1

```python
#Muestra de audio original.
sample_rate, data = wavfile.read('./audios/sine-mono.wav')
#Duracion del archivo original de audio.
duration = 1.0 
time = np.arange(0, duration, 1/sample_rate)
silence = np.zeros(sample_rate*int(duration))

#Limites superior e inferior.
LS = 13400
LI = -10000
#realizar clipping the audio


for i in range(0,data.size):
    if data[i] > LS:
        data[i] = LS
    if data[i] < LI:
        data[i] = LI

#Reproducir audio original.
#display(Audio('./audios/SINE.wav', autoplay=False))

#dibujar audio
plot.plot(time, data[:len(time)])

#Ajustar escala en X\
plot.xlim(0.3, 0.306)
plot.ylim(-16000,16000)
plot.title('Seno con clipping')
plot.show()


#Reproducir audio modificado.
display(Audio(data, rate=sample_rate, autoplay=False))

```

### Grabacion 2

```python
#Muestra de audio original.
sample_rate, data = wavfile.read('./audios/masterofpuppets-mono.wav')
#Duracion del archivo original de audio.
duration = 7 
time = np.arange(0, duration, 1/sample_rate)
silence = np.zeros(sample_rate*int(duration))

#Limites superior e inferior.
LS = 2200
LI = -1800
#realizar clipping the audio


for i in range(0,data.size):
    if data[i] > LS:
        data[i] = LS
    if data[i] < LI:
        data[i] = LI

#Reproducir audio original.
#display(Audio('./audios/SINE.wav', autoplay=False))

#dibujar audio
plot.plot(time, data[:len(time)], color='green')

#Ajustar escala en X\
plot.xlim(1, 1.034 )
plot.ylim(-5000,5000)
plot.title('Grabacion 2 con clipping')
plot.xlabel('Tiempo (s)')
plot.ylabel('Amplitud')
plot.show()


#Reproducir audio modificado.
display(Audio(data, rate=sample_rate, autoplay=False))
```
