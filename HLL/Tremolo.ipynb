```python
#Tremolo effect
"""
Implementacion fuera de linea del efecto tremolo en un archivo de audio.

"""

import numpy as np
import matplotlib.pyplot as plot
from scipy.io import wavfile
from IPython.display import Audio, display
```

```python
#Muestra de audio original.
sample_rate, data = wavfile.read('./audios/sine-mono.wav')
#Duracion del archivo original de audio.
duration = 1.0 
time = np.arange(0, duration, 1/sample_rate)
silence = np.zeros(sample_rate*int(duration))

#Reproducir audio original.
display(Audio('./audios/SINE.wav', autoplay=False))

#dibujar audio
plot.plot(time, data[:len(time)])

#Ajustar escala en X\
plot.xlim(0, 1.0)
plot.ylim(-16000,16000)
plot.xlabel('Tiempo [s]')
plot.ylabel('Amplitud')
plot.title('Grabacion 1. Sin alteraciones') 
plot.show()


```

## Resultado del efecto

### Grabacion 1

```python
#Muestra de audio.
sample_rate, data = wavfile.read('./audios/sine-mono.wav')
#Duracion del archivo original de audio.
duration = 1.0 
indice_modulacion = 1.0
time = np.arange(0, duration, 1.0/sample_rate)


#Frecuencias de tremolo (Hz)
f1 = 100
#Definir la señal de tremolo
tremolo = np.sin(2*np.pi*f1*time) 

#Aplicar efecto tremolo al modular por amplitud las dos señales.
data = (1+indice_modulacion*tremolo) * data

#Reproducir audio modificado.
display(Audio(data, rate=sample_rate, autoplay=False))

#dibujar audio
plot.plot(time, data[:len(time)])

#Ajustar escala en X\
plot.xlim(0, 0.05)

plot.title('Grabacion 1 con Tremolo')
plot.xlabel('Tiempo (s)')
plot.ylabel('Amplitud')
plot.show()
```
